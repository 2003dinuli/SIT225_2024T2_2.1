```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
df = pd.read_csv('hc_sr04_data.csv', header=None, names=['Timestamp', 'Distance'])
print(df.head())  # Print the first few rows of the DataFrame

# Convert Timestamp column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y%m%d%H%M%S')
print(df.head())  # Print the DataFrame again to check the Timestamp conversion

# Clean the Distance column to extract numeric values
df['Distance'] = df['Distance'].str.extract('(\d+)', expand=False)  # Extract numeric part from string
df['Distance'] = pd.to_numeric(df['Distance'], errors='coerce')  # Convert to numeric

print(df.head())  # Print the DataFrame again to check the Distance conversion

# Plot Distance data
plt.figure(figsize=(10, 5))
plt.plot(df['Timestamp'], df['Distance'], marker='o', linestyle='-', color='r')
plt.xlabel('Timestamp')
plt.ylabel('Distance (cm)')
plt.title('HC-SR04 Distance Data Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
